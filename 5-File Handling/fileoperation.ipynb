```python
#Writing  a file(With overriding) 
with open('example.txt','w') as file:
    file.write('Hello World!\n')
    file.write('This will overwrite the existing content.')
```

```python
#Writing  a file(Without overriding) 
with open('example.txt','a') as file:    #append
    file.write('\nHello World!\n')
    file.write('This will write without overwrite the existing content.')
```

```python
#Read a text file and count the number of character,words and lines.
def count_text_file(file_path):
    with open(file_path,'r') as file:
        lines=file.readlines()
        line_count=len(lines)
        word_count=sum(len(line.split()) for line in lines)
        char_count=sum(len(line) for line in lines)
    return line_count,word_count,char_count

file_path='example.txt'
lines,words,characters=count_text_file(file_path)
print(f'Lines:{lines},Words:{words},Characters:{characters}')
```

```python
#Writing and then reading a file
with open('example1.txt','w+') as file:       #w+  : both reading and writing.If file doesn't exist=>created, If exist content will be overwritten.
    file.write("Hello World!\n")
    file.write("THis is new line\n")

    file.seek(0)    # Move the cursor pointer to the top

    content=file.read()
    print(content)
```
