```python
#Iterator-Allow for efficient looping and memory management(without exposing the underlying structure)
mylist=list([1,2,3,4,5,6])
print(type(mylist))
```

```python
myit=iter(mylist)
print(type(myit))
next(myit)
```

```python
try:
    print(next(myit))
except StopIteration:
    print("End of iterator reached")
```
