```python
#Closures:Function inside function(local variable create in outer function can also be access in inner function)
def main_welcome():
    msg="Welcome"
    def sub_welcome_method():
        print("Please understand the concept clearly")
        print(msg)
        print("Enjoy your learning")
    return sub_welcome_method()
```

```python
main_welcome()
```

```python
def main_welcome1(func,arr):
    def sub_welcome_method():
        print("Please understand the concept clearly")
        print(func(arr))
        print("Enjoy your learning")
    return sub_welcome_method()
```

```python
main_welcome1(len,["a","b","c"])
```

```python
##Decorator
def main_welcome(func):
    msg="Welcome"
    def sub_welcome_method():
        print("Please understand the concept clearly")
        print(msg)
        func()
        print("Enjoy your learning")
    return sub_welcome_method()
```

```python
def intro():
    print("This course is related to python")

intro()
```

```python
main_welcome(intro)
```

```python
@main_welcome
def intro():
    print("This course is related to python")
```

```python
#Another example of decorator
def my_decorator(func):
    def wrapper():
        print("Before executing the function")
        func()
        print("After executing the function")
    return wrapper

@my_decorator
def say_hello():
    print("Hello!")

say_hello()
```

```python
#Decorator with arguments
def repeat(n):
    def decorator(func):
        def wrapper(*args,**kwargs):
            for _ in range(n):
                func(*args,**kwargs)
        return wrapper
    return decorator

@repeat(3)
def say_hello():
    print("Hello!")

say_hello()
```

### Decorator provide a clean and readable way to add functionality such as logging, access control, memoization,and more to existing functions without  modifying their code directly.
