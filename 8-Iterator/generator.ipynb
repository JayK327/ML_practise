```python
#Generators : Its simple way to create iterators.They use a yield keyword to produce a series of values lazily(generate value on the fly,doesn't store in memory)
```

```python
def square(n):
    for i in range(3):
        return i**2

print(square(3))  
```

```python
def square_1(n):
    for i in range(3):
        yield i**2

for i in square_1(3):
    print(i)
```

```python
a=square_1(3)
```

```python
print(next(a))
```

```python
def my_generator():
    yield 1
    yield 2
    yield 3
    
gen=my_generator()
```

```python
print(next(gen))
```

```python
for val in my_generator():
    print(val)
```

Practical example-Reading large files(They allow us to process one line at a time without loading the entire file in memory).

```python
#Reading large files
def read_file_line_by_line(file_path):
    with open(file_path,'r') as file:
        for line in file:
            yield line

```

```python
file='large_file.txt'
for line in read_file_line_by_line(file):
    print(line.strip())
    print('---')
```
