```python
# 5 point summary, Box Plot
#Minimum, Maximum, Q1, Q3 ,Median, IQR
import numpy as np
import random
# marks = [random.randint(0, 100) for _ in range(20)]
marks=[9, 61, 86, 77, 70, 23, 96, 55, 86, 12, 49, 52, 97, 51, 20, 67, 49, 24, 88, 24]
#Minimum,Q1,Median,Q3,Maximum=np.quantile(marks,[0,0.25,0.5,0.75,1])
```

```python
q1=np.quantile(marks,0.25)
q3=np.quantile(marks,0.75)
median=np.median(marks)
iqr=q3-q1
lower_bound=q1-1.5*iqr
upper_bound=q3+1.5*iqr
outliers_marks=[mark for mark in marks if mark<lower_bound or mark>upper_bound]
print(lower_bound,upper_bound)
print(outliers_marks)
```

```python
import seaborn as sns
sns.boxplot(marks)  
```
