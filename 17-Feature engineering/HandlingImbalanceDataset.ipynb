```python
import numpy as np 
import pandas as pd
np.random.seed(0)

n_samples = 1000
class_0_ratio=0.9
n_class_0 = int(n_samples * class_0_ratio)
n_class_1 = n_samples - n_class_0
```

```python
class_0=pd.DataFrame({
    'feature_1': np.random.normal(loc=0.0, scale=1.0, size=n_class_0),  #loc: mean, scale: stddev, size: number of samples
    'feature_2': np.random.normal(loc=0.0, scale=1.0, size=n_class_0),
    'label': np.zeros(n_class_0, dtype=int)
})
class_1=pd.DataFrame({
    'feature_1': np.random.normal(loc=2.0, scale=1.0, size=n_class_1),
    'feature_2': np.random.normal(loc=2.0, scale=1.0, size=n_class_1),
    'label': np.ones(n_class_1, dtype=int)
})
df=pd.concat([class_0, class_1], ignore_index=True)
#df=pd.concat([class_0, class_1]).reset_index(drop=True)
```

```python
df
```

```python
df['label'].value_counts()
```

# UPSAMPLING


```python
df_minority = df[df['label'] == 1] 
df_majority = df[df['label'] == 0]
```

```python
from sklearn.utils import resample
df_minority_upsampled=resample(df_minority,replace=True,n_samples=len(df_majority),random_state=42) #replace=True:sampling with replacement
```

```python
df_minority_upsampled.shape
```

```python
df_upsampled= pd.concat([df_majority, df_minority_upsampled])
```

```python
df_upsampled.shape
```

```python
df_upsampled['label'].value_counts()
```

# DOWNSAMPLING


```python
df_minority = df[df['label'] == 1] 
df_majority = df[df['label'] == 0]
```

```python
from sklearn.utils import resample
df_majority_downsampled=resample(df_majority,replace=False,n_samples=len(df_minority),random_state=42) #replace=True:sampling with replacement
```

```python
df_majority_downsampled.shape
```

```python
df_downsampled= pd.concat([df_majority_downsampled, df_minority])
df.shape
```
