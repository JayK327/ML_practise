```python
#try ,except ,else block: If no exception is thrown then after try block ,else block is executed.
try:
    num=int(input("Enter a number"))
    result=10/num
except ValueError as ex:
    print(ex)
except ZeroDivisionError as ex1:
    print(ex1)
except Exception as ex2:
    print(ex2)
else:
    print(f'The result is: {result}')
```

```python
#try ,except ,else and finally: finally will execute whether you exception came or not.
try:
    num=int(input("Enter a number"))
    result=10/num
except ValueError as ex:
    print(ex)
except ZeroDivisionError as ex1:
    print(ex1)
except Exception as ex2:
    print(ex2)
else:
    print(f'The result is: {result}')
finally:
    print("I will execute whether you exception came or not")
```

```python
try:
    file= open('example.txt','r')
    content=file.read()
    print(content)
except FileNotFoundError as ex:
    print(ex)
finally:
    if 'file' in locals():
        if not file.closed:
            file.close()
            print("File closed")
```
