```python
# Creating a in memory database
import sqlite3
```

```python
connection= sqlite3.connect('example.db')
connection
```

```python
cursor=connection.cursor()
```

```python
#Create a table 
cursor.execute('''
 CREATE TABLE IF NOT EXISTS employees(
               id INTEGER PRIMARY KEY,
               name TEXT NOT NULL,
               age INTEGER,
               department TEXT
               )
''')

#Commit the changes
connection.commit()
```

```python
cursor.execute('''
SELECT * FROM employees    
''')
```

```python
cursor.execute('''
Insert into employees(name,age,department)
               values ('ABC',30,'IT')
''')
```

```python
connection.commit()
#Fetch the data
cursor.execute('''
SELECT * FROM employees
''')
rows=cursor.fetchall()
for row in rows:
    print(row)
```
