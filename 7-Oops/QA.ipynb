### 2.How super() Works with Multiple Inheritance

In multiple inheritance, when you call super(), Python follows the MRO to determine which classâ€™s method to call next. The MRO is determined by the C3 Linearization algorithm, which ensures that the order of method calls follows the hierarchy defined in the inheritance chain.


```python
class Person:
    def __init__(self, name):
        self.name = name

class Employee:
    def __init__(self, employee_id):
        self.employee_id = employee_id

class Manager(Person, Employee):
    def __init__(self, name, employee_id):
        super().__init__(name)  # Calls Person's __init__
        Employee.__init__(self, employee_id)  # Directly calls Employee's __init__
```
