## Magic Method
#### Magic methods in python also known as dunder methods(double underscore methods).These methods enable you to define the benefit of objects for built-in operation such as arithmetic n,comparision and more. (eg.  __init__ ,__str__, __len__).

```python
class Person:
    def __init__(self, name, age):
        self.name = name
        self.age = age

    def __str__(self):
        return f"Person(Name: {self.name}, Age: {self.age})"

person=Person("Rahul",54)
print(person.__str__())
```

### Operator Overloading


```python
# __add__, __sub__,__lt__,__gt__, __eq__
class Vector:
    def __init__(self, x, y):
        self.x = x
        self.y = y

    def __add__(self, other):
        return Vector(self.x + other.x, self.y + other.y)

    def __sub__(self, other):
        return Vector(self.x - other.x, self.y - other.y)

    def __lt__(self, other):
        return (self.x**2 + self.y**2) < (other.x**2 + other.y**2)

    def __gt__(self, other):
        return (self.x**2 + self.y**2) > (other.x**2 + other.y**2)

    def __eq__(self, other):
        return self.x == other.x and self.y == other.y

    def __str__(self):
        return f"Vector({self.x}, {self.y})"
    
v1=Vector(2,3)
v2=Vector(5,6)
v3=v1+v2
print(v3)  
```
