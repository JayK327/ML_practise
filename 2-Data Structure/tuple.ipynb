# Tuple
### Ordered collection of item but immutable,can be of different datatype

```python
empty_tuple=()
print(empty_tuple)
print(type(empty_tuple))
```

```python
lst=list()
tpl=tuple()
print(type(lst))
print(type(tpl))
```

### -Slicing and indexing are same as list

#### Operations like tpl*3 or tpl+tpl is allowed

```python
numbers=(1,2,3,4,5,6)
print(numbers+numbers)
print(numbers*3)
print(type(numbers))
print(numbers.count(1))
print(numbers.index(3))
```

### -Packing and unpacking tuple

```python
packed_tuple=1,"Hello",3.14     #By default it takes as tuple
print(packed_tuple)
a,b,c=packed_tuple
print(a)
print(b)
print(c)
```

```python
#Unpacking with *
numbers=1,2,3,4,5,6
first,*middle,last=numbers
print(first)
print(middle)
print(last)

```
