
### Dictionary are unordered collection of items.Stores in key,value pair .Key must be unique and immutable.

```python
empty_dict={}
print(type(empty_dict))
empty_dict1=dict()
print(empty_dict1)
```

```python
student={"name":"Raj","age":24,"name":26}
print(student)
print(student['age'])  ##alternete using grade student.get('grade')
print(student.get('grade'))
print(student.get('grade',"Not available"))  #Giving default value 
```

```python
del student['age']
print(student)
```

### Dictionary methods like d.keys(),d.values(),d.items()

```python
student={"name":"Karan","age":26,"country":"India"}
print(student)
```

```python
#Deep copy
student_copy=student
print(student_copy)
print(student)
```

```python
student['name']="Rahul"
print(student_copy)
print(student)
```

```python
##Shallow copy
student_copy_1=student.copy()
student['name']="Karan"
print("student_copy_1:" ,student_copy_1)
print("student:" ,student)
```

```python
squares={x:x**2 for x in range(5)}
print(squares)
```

```python
##Merged two dictionaries into one
dict1={"a":1,"b":2}
dict2={"b":3,"c":4}
dict={**dict1,**dict2}
print(dict)
```
